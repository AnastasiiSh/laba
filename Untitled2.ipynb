```python
html = 'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating'
arr_links = [html]

import requests
import re
from urllib.request import urlopen
import urllib
import urllib
import csv

import pandas as pd
def write_to_csv(self,output):
	df_cont1 =pd.DataFrame(columns=['Название'])
	df_cont2 =pd.DataFrame(columns=['Год'])
	df_cont3 =pd.DataFrame(columns=['Время'])
	df_cont4 =pd.DataFrame(columns=['Жанр'])
	df_cont5 =pd.DataFrame(columns=['Рейтинг'])

	#df_output =  pd.DataFrame(output,columns=['Index','Name'],dtype=float)
	df_output.to_csv('output_films.csv')

names_csv = []
year_csv = []
time_csv = []
genre_csv = []
stars_csv = []
relations_table = []

#write_to_csv([['1','IBM'],['2','Cogni'],['3','Toyota'],['4','tomtom']])

for check_link in arr_links:
	mask1 = r'\>'
	mask2 = r'\<'
	mask11 = r'\('
	mask22 = r'\)'
	r = urlopen(check_link)
	source_code = r.read()
	soup = BeautifulSoup(source_code, 'html.parser')
	#print(len(soup.find_all('h3', class_='lister-item-header')))
	#print(soup.find_all('h3', class_='lister-item-header'))
# НАЗВАНИЯ ФИЛЬМОВ
	table = soup.find_all('h3', class_='lister-item-header')
	for i in table:
	#my_table = table[:]
		name = i.find('a')
# 		print(str(name))
		ch1 = [k.start() for k in re.finditer(mask1, str(name))][0]
		ch2 = [k.start() for k in re.finditer(mask2, str(name))][1]
		name = str(name)
		name = name[ch1+1:ch2]
		names_csv.append(name) 
		print(str(name))
	relations_table.append(names_csv)
# ГОД ФИЛЬМА
	table = soup.find_all('h3', class_='lister-item-header')
	for i in table:
	#my_table = table[:]
		tr = i.find('span', class_='lister-item-year text-muted unbold')
		ch1 = [k.start() for k in re.finditer(mask11, str(tr))][0]
		ch2 = [k.start() for k in re.finditer(mask22, str(tr))][0]
		tr = str(tr)
		tr = tr[ch1+1:ch2]
		year_csv.append(tr)         
		print(str(tr))
	relations_table.append(year_csv)
# ВРЕМЯ ФИЛЬМА
	table = soup.find_all('p', class_='text-muted')
	for i in table:
	#my_table = table[:]
		tr = i.find('span', class_='runtime')
		if str(tr) != 'None':
			ch1 = [k.start() for k in re.finditer(mask1, str(tr))][0]
			ch2 = [k.start() for k in re.finditer(mask2, str(tr))][1]
			tr = str(tr)
			tr = tr[ch1+1:ch2]
			time_csv.append(tr)
			print(str(tr))
	relations_table.append(time_csv)            
# ЖАНР ФИЛЬМА
	table = soup.find_all('p', class_='text-muted')
	for i in table:
	#my_table = table[:]
		tr = i.find('span', class_='genre')
		if str(tr) != 'None':
			ch1 = [k.start() for k in re.finditer(mask1, str(tr))][0]
			ch2 = [k.start() for k in re.finditer(mask2, str(tr))][1]
			tr = str(tr)
			tr = tr[ch1+1:ch2].strip()
			genre_csv.append(tr)
			print(str(tr).strip())
	relations_table.append(genre_csv)
# РЕЙТИНГ ФИЛЬМА
	table = soup.find_all('div', class_='inline-block ratings-imdb-rating')
	for i in table:
	#my_table = table[:]
		tr = i.find('strong')
		ch1 = [k.start() for k in re.finditer(mask1, str(tr))][0]
		ch2 = [k.start() for k in re.finditer(mask2, str(tr))][1]
		tr = str(tr)
		tr = tr[ch1+1:ch2]
		stars_csv.append(tr)        
		print(str(tr))
	relations_table.append(stars_csv)
	relations_table2 = []
	relations_table3 = []
# 	for un_num1, un_item1 in enumerate(relations_table):
# 		for un_num2, un_item2 in enumerate(un_item1):       
	for un_num in range(len(relations_table[0])):
		for un_num2, un_item in enumerate(relations_table):
			if un_num2 == un_num:
				relations_table2.append(un_item)
		#relations_table3.append(relations_table2)
print(str(relations_table2))

```
